```python
#define Dense model (MLP)
def denseModel(inputLen, outputLen):
    model = tf.keras.models.Sequential([
      tf.keras.layers.Dense(1000, activation='relu', input_shape=(inputLen,)),
      tf.keras.layers.Dense( 500, activation='relu'),
      #model.add(tf.keras.layers.Dense( 100, activation='relu'))
      tf.keras.layers.Dense(outputLen, activation='softmax')
    ])
    return model
```

```python
#Define CNN model 1
def cnnModel1(inputLen, inputCols=1, outputLen=1, dropOutRatio=0.5, filters=10, ksize=20):
    model = tf.keras.models.Sequential([
    
    # (batch, 50000, 1) --> (batch, 10000, 1)
    tf.keras.layers.Conv1D(
        #filters=100, kernel_size=10, 
        filters=filters, kernel_size=ksize,   #10 seems to have the same effect
        strides=1, padding='same', activation = tf.nn.relu, 
        input_shape=(inputLen, inputCols, ))
        ,
    tf.keras.layers.MaxPooling1D(
        #pool_size=10, strides=5, padding='same')
        #pool_size=10, strides=10, padding='same') #seems to loose accuracy however is faster
        pool_size=5, strides=5, padding='same') #5 seems to have the same effect as 10/5
        ,
    # (batch, 5000, 1) --> (batch, 500, 1)
    tf.keras.layers.Conv1D(
        #filters=100, kernel_size=10, 
        filters=filters, kernel_size=ksize,   #10 seems to have the same effect
        strides=1, padding='same', activation = tf.nn.relu)
        ,
    tf.keras.layers.MaxPooling1D(
        pool_size=10, strides=10, padding='same')
        ,
    # (batch, 500, 1) --> (batch, 50, 1)
    tf.keras.layers.Conv1D(
        #filters=100, kernel_size=10, 
        filters=filters, kernel_size=ksize, 
        strides=1, padding='same', activation = tf.nn.relu)
        ,
    tf.keras.layers.MaxPooling1D(
        pool_size=10, strides=10, padding='same')
        ,
    
    # adding a dropout layer for the regularization and avoiding over fitting
    tf.keras.layers.Dropout(dropOutRatio)
        ,
    #Flatten
    tf.keras.layers.Flatten()
        ,
    #classifier
    #DOES NOT improve accuracy
    #tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(outputLen, activation='softmax'),
    ])
    return model
```
